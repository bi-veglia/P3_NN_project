```python
def split_list(a_list):
    half = len(a_list)//2
    return a_list[:half], a_list[half:]
```

```python
bpmoff=['BPM_NR_19','BPM_NOR_52','BPM_NOR_98','BPM_OL_144','BPM_OL_98','BPM_OL_75','BPM_SWL_31']
directory = 'data/data_matlab/'
```

```python
for filename in os.scandir(directory):
    mat = scipy.io.loadmat(filename)
    b=filename.name.split('_')
    a=(pd.Series(mat)).to_frame()
    a.drop(['__header__','__version__','__globals__','t'])
    bpms=[]
    for i in a[0]['bpmnames']:
        bpms.append(i[0][0])
    gaps=a[0]['gap'][0]
    _,gaps=split_list(gaps)
    x=a[0]['x'][:,:len(gaps)].transpose()
    y=a[0]['y'][:,:len(gaps)].transpose()
    ax=pd.DataFrame(data=x,index=gaps,columns=bpms)
    ay=pd.DataFrame(data=y,index=gaps,columns=bpms)
    ax=ax.drop(columns=bpmoff)
    ax = ax.add_suffix('_x')
    ay=ay.drop(columns=bpmoff)
    ay = ay.add_suffix('_y')
    axy=ax.join(ay)
    file_name='data/data_csv/'+str(b[1])+'_xy.csv'
    axy.to_csv(file_name)
    #file_namey='data/data_csv/'+str(b[1])+'_y.csv'
    #ax.to_csv(file_namex)
    #ay.to_csv(file_namey)
    #globals()[str(b[1])+'_dfy'].to_csv('data/data_csv')
```

```python
i=0
for filename in os.scandir('data/data_csv/'):
    if filename.is_file():
        df = pd.read_csv(filename,index_col=0)
        df1=df.loc[~df.index.duplicated(), :]
        indexList = np.linspace(df.index.min(),df.index.max(),709-len(df1.index))
        if filename.name.startswith('pu04_'):
            df2=df1
        else:  
            df2 = df1.reindex(df1.index.union(indexList)).interpolate('index')
        df2.reset_index(inplace=True)
        df2 = df2.rename(columns = {'index':'gap'+str(i)})
        print(df.index.max())
        if i==0:
            df2_old=df2.copy()
        else:
            df3=pd.concat([df2_old,df2], ignore_index = True,sort = False)
            #df3['gap'+str(i-1)] = df3['gap'+str(i-1)].fillna(df.index.max())
            df2_old=df3.copy()
        i+=1
```
